```python
import pandas as pd
import csv

state_pop = {}
with open('pop.csv') as f:
    
    lines = f.readlines()
    
    for i in range(6,len(lines)):
        line = lines[i].strip().split(",")
        s = us_state_abbrev[line[4]]
        state_pop[s] = int(line[15])  
state_pop    
    
```

```python
hos = {}

df = pd.read_csv('h.csv')

for s in list(df["State"]):
    if s not in hos:
        hos[s] = 1
    else:
        hos[s]+=1
hos
```

```python
fin = {}
for k,v in state_pop.items():

    fin[k] = v//hos[k]
fin = sorted(fin.items(), key=lambda d: d[1], reverse=False)
fin = dict(fin)
fin
```

```python
import matplotlib.pyplot as plt

new_dic = {"s":[],"v":[]}
for k,v in fin.items():
    new_dic["s"].append(k)
    new_dic["v"].append(v)
    
df = pd.DataFrame(new_dic, index = fin.keys())

fig,ax = plt.subplots(figsize = (10,6))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("How Many People Share One Hospital In Every State",size=20)
ax.set_xlabel("State",size=18)
ax.set_ylabel("How Many People Share One Hospital",size=18)

df[-10:].plot.barh(ax = ax, legend = False,color="black")

new_df = df[-10:]
new_df = new_df.reset_index()

new_df

for i in range(len(new_df)):
    ax.text(new_df.loc[i,"v"]+1000,i,str(new_df.loc[i,"v"]))
```

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
import pandas as pd
```

```python
us = gpd.read_file("zip://states.zip")
```

```python
us.set_index("STUSPS", drop=False, inplace=True)
continent = us[~us["STUSPS"].isin(["AK", "HI", "PR"])]
ak = us.loc["AK":"AK"]
hi = us.loc["HI":"HI"]
pr = us.loc["PR":"PR"]
```

```python
fig = plt.figure(figsize = (10,8))
gs = fig.add_gridspec(5, 3)
ax1 = fig.add_subplot(gs[:3, :])
ax2 = fig.add_subplot(gs[3, 0])
ax3 = fig.add_subplot(gs[3, 1])
ax4 = fig.add_subplot(gs[3, 2])
continent.plot(color="orange", edgecolor="k", ax=ax1)
ak.plot(color="orange", edgecolor="k", ax=ax2)
hi.plot(color="orange", edgecolor="k", ax=ax3)
pr.plot(color="orange", edgecolor="k", ax=ax4)
ax2.set_xlim(-180, -120)
for ax in fig.axes:
    ax.axis("off")
```

```python
pph=pd.DataFrame.from_dict(fin, orient = "index",columns = ["pph"])
```

```python
pph.head()
```

```python
for rows in pph.index:
    pph.loc[rows,"color"] ="lightsalmon"
pph.head()
```

```python
for rows in pph.index:
    if pph.loc[rows,"pph"] >= 30000:
        pph.loc[rows,"color"] = "tomato"
    if pph.loc[rows,"pph"] >= 60000:
        pph.loc[rows,"color"] = "red"
    if pph.loc[rows,"pph"] >= 90000:
        pph.loc[rows,"color"] = "brown"
```

```python
for states in continent.index:
    continent.loc[states,"color"] = pph.loc[states,"color"]
```

```python
ak.loc["AK","color"] = pph.loc["AK","color"]
pr.loc["PR","color"] = pph.loc["PR","color"]
hi.loc["HI","color"] = pph.loc["HI","color"]
```

```python
import matplotlib.patches as mpatches

fig = plt.figure(figsize = (16,12))


red_patch = mpatches.Patch(color='red', label='The red data')
plt.legend(handles=[red_patch])


gs = fig.add_gridspec(5, 3)
ax1 = fig.add_subplot(gs[:3, :])
ax2 = fig.add_subplot(gs[3, 0])
ax3 = fig.add_subplot(gs[3, 1])
ax4 = fig.add_subplot(gs[3, 2])
continent.plot(color=continent["color"], edgecolor="k", ax=ax1,legend = True)
ak.plot(color=ak["color"], edgecolor="k", ax=ax2)
hi.plot(color=hi["color"], edgecolor="k", ax=ax3)
pr.plot(color=pr["color"], edgecolor="k", ax=ax4)
ax2.set_xlim(-180, -120)
for ax in fig.axes:
    ax.axis("off")
```
